```python
plt.figure(figsize=(12,8))
df['Age'].dropna().hist(bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
def convert_age(value):
    if pd.isna(value):
        return np.nan
    
    # Handle specific labels
    if value.startswith("Under"):
        return 15  # guess midpoint of <18
    if value.startswith("65"):
        return 70  # midpoint of 65+
    if value.startswith("Prefer"):
        return np.nan
    
    # Handle ranges like "25-34 years old"
    try:
        low, high = value.split(" ")[0].split("-")
        low = int(low)
        high = int(high)
        return (low + high) / 2
    except:
        return np.nan

df["Age_numeric"] = df["Age"].apply(convert_age)
df["Age_numeric"].skew()
```

```python
plt.figure(figsize=(6,4))
df['Age_numeric'].plot(kind='box')
plt.title('Age Boxplot')
plt.show()
```
