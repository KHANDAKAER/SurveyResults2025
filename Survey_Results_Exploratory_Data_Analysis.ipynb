```python
#converting string to numeric (middle age chosen from each range)

def convert_age(value):
    if pd.isna(value):
        return np.nan
    
    # Handle specific labels
    if value.startswith("Under"):
        return 15  # guess midpoint of <18
    if value.startswith("65"):
        return 70  # midpoint of 65+
    if value.startswith("Prefer"):
        return np.nan
    
    # Handle ranges like "25-34 years old"
    try:
        low, high = value.split(" ")[0].split("-")
        low = int(low)
        high = int(high)
        return (low + high) / 2
    except:
        return np.nan

df["Age_numeric"] = df["Age"].apply(convert_age)
df["Age_numeric"].skew()
```

```python
#import libraries for PCA

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
#only numeric columns

numeric_df = df.select_dtypes(include=['int64', 'float64'])
numeric_df.head()
```

```python
#Drop rows with missing values

numeric_df = numeric_df.dropna()
numeric_df.shape
```

```python
#Standardize the numeric data

scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_df)
```

```python
#PCA

pca = PCA()
pca_components = pca.fit_transform(scaled_data)
```

```python
#how much each component explains variance

plt.figure(figsize=(10,5))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.title('Cumulative Explained Variance by PCA Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid()
plt.show()
```

```python
#which features matter most

loadings = pd.DataFrame(
    pca.components_.T,
    columns=[f'PC{i+1}' for i in range(pca.n_components_)],
    index=numeric_df.columns
)
loadings.head()
```

```python
#Visualize the first two principal components

plt.figure(figsize=(8,6))
plt.scatter(pca_components[:, 0], pca_components[:, 1], alpha=0.5)

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA: PC1 vs PC2')
plt.show()
```

```python
# Machine Learning Pipeline

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df['YearsCode_numeric'] = df['YearsCode']
```

```python
# select target variable "AI_binary"
# From AISelect, treat any kind of “Yes, I use AI tools…” as 1, and “No…” as 0.
ai_map = {
    "Yes, I use AI tools daily": 1,
    "Yes, I use AI tools weekly": 1,
    "Yes, I use AI tools monthly or infrequently": 1,
    "No, but I plan to soon": 0,
    "No, and I don't plan to": 0,
}

df['AI_binary'] = df['AISelect'].map(ai_map)
```

```python
#just drop the rows with NA

df = df.dropna(subset=['EdLevel', 'Employment'])
model_df = df.dropna(subset=['Age_numeric', 'YearsCode_numeric', 'AI_binary'])
model_df.shape
```

```python
features = ['Age_numeric', 'YearsCode_numeric', 'Employment', 'EdLevel']

X = model_df[features]
y = model_df['AI_binary']
```

```python
#One-hot encode categorical columns
#Employment and EdLevel are categorical, so we need dummy variables.

X_encoded = pd.get_dummies(X, drop_first=True)
X_encoded.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y,
    test_size=0.2,
    random_state=42
)

X_train.shape, X_test.shape
```

```python
#Train the model with Random Forest

model = RandomForestClassifier(
    n_estimators=300,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
#Evaluate the model

y_pred = model.predict(X_test)

acc  = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec  = recall_score(y_test, y_pred)
f1   = f1_score(y_test, y_pred)

print("Accuracy :", acc)
print("Precision:", prec)
print("Recall   :", rec)
print("F1-score :", f1)
```

```python
#Confusion matrix

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix: AI Usage Prediction')
plt.show()
```

```python
#Feature importance

importances = pd.Series(model.feature_importances_, index=X_encoded.columns)
importances.sort_values(ascending=False).head(10)
```
